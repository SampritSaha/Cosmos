```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("phl_exoplanet_catalog_2019.csv")
```

```python
data.head()
```

```python
#finding range of data
```

```python
max_col=data.max(axis='index')
min_col=data.min(axis='index')

print(max_col)
print(min_col)
```

```python
#code to make a table of max and mean
```

```python
#finding mean of each row
```

```python
mean_col = data.mean(axis='index')
print("Mean:\n\n",mean_col)
```

```python
#finding median of data
```

```python
median_row = data.median(axis='index')
print("Median: \n\n",median_row)
```

```python
#finding standerd deviation of data
```
